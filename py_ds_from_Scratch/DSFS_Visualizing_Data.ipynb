```python
from matplotlib import pyplot as plt
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
```

```python
plt.title('Nominal GDP')
```

```python
plt.ylabel("Billions of $")
```

```python
plt.show()
```

```python
# Bar chart
movies = ["Star Wars", "Count of Monte Cristo", "Indiana Jones", "Lord of the Rings", "Harry Potter"]
num_movies = ["8", "1", "5", "3", "8"]
```

```python
# bars are .8 width, add .1 to center
xs = [i + .1 for i, _ in enumerate(movies)]
```

```python
plt.bar(xs, num_movies, 8)
```

```python
plt.ylabel("# of Movies in Series")
plt.title("My Favorite Series")
```

```python
# label x axis with movie names at bar centers
plt.xticks([i + .5 for i, _ in enumerate(movies)], movies)
```

```python
plt.show()
```
