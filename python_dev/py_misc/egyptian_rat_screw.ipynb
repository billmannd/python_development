```python
from random import shuffle
suits = ["Spades","Clubs","Hearts","Diamonds"]
card = ["2","3","4","5","6","7","8","9","10","Jack","Queen","King","Ace"]
deck = []
```

```python
for i in suits:
    for j in card:
        deck.append({i:j})
```

```python
def get_players(num_players):
    """
    
    num_players: integer variable for how may players there are for the game
    """
    
    if num_players > 20:
        return "Too many people"
    
    players = []
    for i in range(1,num_players+1):
        players.append([])
    return players    
    
```
