# Regression Plots

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill', y = 'tip', data =tips, hue='sex', markers=['o','v'], scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', col='day',aspect=0.6,size=8)
```
