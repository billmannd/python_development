```python
import matplotlib.pyplot as plt 
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y,'g-')
plt.xlabel("Iggy")
plt.ylabel('Nikki')
plt.title('Big Booty')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g-')
plt.subplot(1,2,2)
plt.plot(y,x,'b-')
```

```python
fig = plt.figure()

axes = fig.add_axes([1,1,.8,.8])
axes.plot(x,y)
axes.set_xlabel('Age')
axes.set_ylabel('Wisdom')
axes.set_title('Maturing')
```

```python
fig = plt.figure()

axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis2 = fig.add_axes([0.2,0.5,0.4,0.3])

axis1.plot(x,y)
axis2.plot(y,x)
axis2.set_title("SMALLER PLOT")
axis1.set_title("LARGER PLOT")
```

```python
#Subplotting: 

#since axes is a list, you can iterate through it
#for current_ax in axes:
 #   current_ax.plot(x,y)

#you can also call each object with list indexes
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)


axes[0].plot(x,y)
axes[0].set_title('First Plot')

axes[1].plot(y,x)
axes[1].set_title('Second Plot')
```

```python
'''Figure Size and DPI'''
```

```python
fig,axes = plt.subplots(nrows = 2, ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
fig.savefig('my_picture.png',dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.plot(x,x**3,label='X Cubed')
ax.plot(x,x**2,label='X Squared')


ax.legend(loc=(0.1,0.8))
```
